```python
time = np.loadtxt('20171003_ch2_interference_w_sas_2.isf', delimiter=',')[:,0]
ch2_interference_w_sas = np.loadtxt('20171003_ch2_interference_w_sas_2.isf', delimiter=',')[:,1]
ch3_interference_w_sas = np.loadtxt('20171003_ch3_interference_w_sas_2.isf', delimiter=',')[:,1]
```

```python
##Defining function to pull desired information from the data
def get_peak_data(array, plotrange):
    peak = []
    maximums = []
    inds = []
    
    for i in range(len(time)):
        if (time[i] >= plotrange[0]) and (time[i] <= plotrange[1]):
            if (array[i-1] < array[i]) and (array[i] < array[i+1]):
                maximums.append(array[i])
                inds.append(i)
            
#    print("Length of peak array data: {}".format(len(peak)))
    print("Maximum(s) found at: t = {}".format([time[i] for i in inds]))
    print("Maximum value(s) occurs at index: {}".format(inds))
    print("Maximum value(s): {}".format(maximums))
    plt.plot([time[i] for i in inds], maximums, 'rp')
    plt.plot(peak)
    return peak
```
