```python
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.signal import argrelmax as arm
import numpy as np
from scipy.optimize import curve_fit
```

```python
##Load in the data from the text file
data = np.loadtxt('20171017_cavendish_driven_osc_1.txt', delimiter = '\t')

##Use scipy built in method to isolate relative maxima and minima
max_inds = arm(data[:,1])

##Find the index corrsesponding to, roughly, when the "constant" amplitude oscialltion begins
cutoff = 0

for i in range(data.shape[0]):
    if data[i,0] >= 1500:
        cutoff = i
        break
```

```python
##Data and plotting for isolation of maxima and minima
max_times = data[max_inds,0]
max_angles = data[max_inds, 1]


plt.plot(data[:,0], abs(data[:,1]))
plt.plot(max_times, abs(max_angles), 'rp');
```

```python
constant_amplitude_times = data[cutoff:, 0]
constant_amplitude_angles = data[cutoff:, 1]

plt.title("Angular Position, Driven Oscillation")
plt.ylabel("Angular Position, Calibrated (mrad)")
plt.xlabel("Time (s)")
plt.plot(constant_amplitude_times, constant_amplitude_angles);
```

```python
constant_amplitude_angles_abs = abs(data[cutoff:, 1])

plt.title("Absolute Value of Angular Position, Driven Oscillation")
plt.ylabel("Angular Position, Calibrated (mrad)")
plt.xlabel("Time (s)")
plt.plot(constant_amplitude_times, constant_amplitude_angles_abs);
```

```python
plt.plot(constant_amplitude_times, constant_amplitude_angles_abs)
plt.plot(constant_amplitude_times[arm(constant_amplitude_angles_abs)], constant_amplitude_angles_abs[arm(constant_amplitude_angles_abs)], 'rp')
```
